```python

# tokens = nltk.word_tokenize(text.lower())
from nltk.util import ngrams
from collections import Counter


lemmatizer = WordNetLemmatizer()
# Apply lemmatization
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]

# Generate trigrams
trigrams = list(ngrams(lemmatized_tokens, 3))

# Count trigram frequency
trigram_freq = Counter(trigrams)

# Function to predict next word
def predict_next_word(word1, word2):
    predictions = {trigram[2]: count for trigram, count in trigram_freq.items() if trigram[0] == word1 and trigram[1] == word2}
    return max(predictions, key=predictions.get) if predictions else None

```

```python
def predict_next_word(w1,w2,trigram_freq):
    result = dict()

    for val in trigram_freq.items():
        item = val[0]
        freq = val[1]

        # cari kelompok yang word1 dan word2 sama dengan input
        if item[0] == w1 and item[1] == w2:
            # tampung hasil jika ketemu dalam dictionary
            result[item] = freq

    # handling ga ketemu
    if not result : 
        return(None)
    else:
        # cari kelompok yang muncul paling banyak
        max_key = max(result, key=lambda k: result[k])
        
        # keluarkan hanya word ke 3 (index ke 2, karena mulai dari 0)
        return(max_key[2])
```

```python
# test schema with random value
import random
iteration_number = 10

def randomizers(iteration_number, max_num):
    list_random = []

    for i in range(iteration_number):
        rand_numbers = random.randint(0, max_num)
        list_random.append(rand_numbers)
    
    return(list_random)

def testing_word_prediction(trigrams, trigram_freq, iteration_number, list_random, detail=False):
    final_score = 0
    
    for i in range (iteration_number):
        score = 0

        # print(trigrams[rand_numbers])
        # value -> trigram -> w1, w2, w3
        # value -> trigram -> value[0], value[1], value[2]
        # w1,w2 (soal), w3(jawaban)
        value = trigrams[list_random[i]]

        # predict next word dengan masukan w1 dan w2
        next_word = predict_next_word(value[0], value[1], trigram_freq)

        if detail:
            print('predict next word: {}'.format(next_word))

        # compare apakah prediksi next word = w3 (jawaban)
        if (next_word == value[2]):
            if detail:
                print('true')
                print()
            
            score = 1
        else :
            if detail:
                print('false')
                print()
        
        #masukan score ke sebuah list final_score 
        final_score += score

    return(final_score)

list_randoms = randomizers(iteration_number, len(trigrams))
final_score = testing_word_prediction(trigrams, trigram_freq, iteration_number, list_randoms)
print('score : {}'.format(final_score))
print('percentage true : {}'.format(final_score*100.00/iteration_number))
```

```python
# testing skenario, dibuat berulang 
def batch_test(trigrams, trigram_freq, batch, iteration_number, rand_list, detail=False):
    list_score = []

    for i in range(batch):
        final_score = testing_word_prediction(trigrams, trigram_freq, iteration_number, rand_list[i], detail)
        list_score.append(final_score)
    
    return(list_score)
```
