```python
lemmatizer = WordNetLemmatizer()
tokens = nltk.word_tokenize(text.lower())

# Apply lemmatization
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

# Generate trigrams
trigrams = list(ngrams(lemmatized_tokens, 3))

# Count trigram frequency
trigram_freq = Counter(trigrams)

# Function to predict next word
def predict_next_word(word1, word2):
    predictions = {trigram[2]: count for trigram, count in trigram_freq.items() if trigram[0] == word1 and trigram[1] == word2}
    return max(predictions, key=predictions.get) if predictions else None

# Example usage
word1, word2 = "i", "love"
next_word = predict_next_word(word1, word2)
print(f"Predicted next word: {next_word}")
```
